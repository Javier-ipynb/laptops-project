# <p style="background-color:#122428;font-family:newtimeroman;color:#1ea4bf;font-size:150%;text-align:center;border-radius:40px 40px;">Importing Libraries</p>

```python
# Data manipulation
import pandas as pd
import numpy as np

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Data pre-processing
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

# Regression algorithms
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.naive_bayes import GaussianNB

# Metrics
from sklearn import metrics
from sklearn.model_selection import cross_val_score

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

# <p style="background-color:#122428;font-family:newtimeroman;color:#1ea4bf;font-size:150%;text-align:center;border-radius:40px 40px;">Data Preprocessing</p>

To begin, we need to encode our data, as the algorithms for training work through matrices of integers. Since we dont have any floating-point values, we only need to encode the columns whose data are of string type.

```python
# Columns to encode
columns = [
    'Model',
    'CPU',
    'Storage type',
    'GPU',
]

# Initialize the label encoder
L_E = LabelEncoder()

# Encode the specified columns
for col in columns:
    df[col] = L_E.fit_transform(df[[col]].values.ravel())
```

Once this is done, we proceed to extract the columns with which we will train our model. This way, each variable will be a vector, and by combining them, we will form a matrix. Then, we will split our data into training and test sets.

```python
# Extract feature arrays from the DataFrame
model = df['Model'].values
cpu = df['CPU'].values
storage_type = df['Storage type'].values
gpu = df['GPU'].values
ram = df['RAM'].values
storage = df['Storage'].values

# Create feature matrix X and target vector y
X = np.array([model, cpu, storage_type, gpu, ram, storage]).T
y = df['Final Price'].values.ravel()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1951)
```

# <p style="background-color:#122428;font-family:newtimeroman;color:#1ea4bf;font-size:100%;text-align:center;border-radius:40px 40px;">Metrics</p>

```python
# Evaluate and print regression metrics
def GeneralMetrics(y_true, y_pred):
    def mean_absolute_percentage_error(y_true, y_pred): 
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    print('Evaluation metric results:-')
    print(f'MSE is : {metrics.mean_squared_error(y_true, y_pred)}')
    print(f'MAE is : {metrics.mean_absolute_error(y_true, y_pred)}')
    print(f'RMSE is : {np.sqrt(metrics.mean_squared_error(y_true, y_pred))}')
    print(f'MAPE is : {mean_absolute_percentage_error(y_true, y_pred)}')
    print(f'R2 is : {metrics.r2_score(y_true, y_pred)}', end='\n\n')
```

# <p style="background-color:#122428;font-family:newtimeroman;color:#1ea4bf;font-size:150%;text-align:center;border-radius:40px 40px;">Model Selection</p>

First, we will test different algorithms to determine which one has the best overall performance. Then, we will select some of the best ones to fine-tune the hyperparameters and achieve the best possible model.

```python
def BasicAlgorithms(X_train, y_train):
    models = {
        'LinearRegression': LinearRegression(n_jobs=-1),
        'SVR': SVR(),
        'DecisionTreeRegressor': DecisionTreeRegressor(random_state=1951),
        'RandomForestRegressor': RandomForestRegressor(random_state=1951, n_jobs=-1),
        'GradientBoostingRegressor': GradientBoostingRegressor(random_state=1951),
        'AdaBoostRegressor': AdaBoostRegressor(random_state=1951),
        'KNeighborsRegressor': KNeighborsRegressor(n_jobs=-1),
        'GaussianNB': GaussianNB()
    }
    
    model_scores = []
    for name, model in models.items():
        # Perform cross-validation
        cv_scores = cross_val_score(model, X_train, y_train, cv=7, scoring='r2')
        
        # Calculate evaluation metrics
        r2 = cv_scores.mean()
        
        # Append the metrics to model_scores
        model_scores.append((name, r2))
    
    # Create a DataFrame from model_scores
    columns = ['Model', 'R2']
    df_model_scores = pd.DataFrame(model_scores, columns=columns)
    
    # Sort the DataFrame by MSE in ascending order
    df_model_scores = df_model_scores.sort_values(by='R2', ascending=False)
    
    # Display the DataFrame
    print(df_model_scores)
```

```python
BasicAlgorithms(X_train, y_train)
```

We found that the algorithms that performed best were: RandomForestRegressor and GradientBoostingRegressor. Therefore, I will create a function for each model, in which we will try to adjust the hyperparameters of each one and select the best final model.

```python
# Train and evaluate a RandomForestRegressor
def RandomForestModel(X_train, y_train, X_test, y_test):
    rf = RandomForestRegressor(random_state=1951)

    # Parameter grid for hyperparameter tuning
    param_grid = {
        'n_estimators': [750, 900, 1200],
        'max_depth': [9, 12, 15],
        'min_samples_split': [3, 5, 7]
    }
    
    # Perform grid search with cross-validation
    rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv=7, n_jobs=-1)

    # Fit the model on training data
    rf.fit(X_train, y_train)
    print(rf.best_params_)
    rf = rf.best_estimator_

    # Predict on test data
    y_pred = rf.predict(X_test)

    # Perform cross-validation to evaluate the model
    cv_scores = cross_val_score(rf, X, y, cv=7, scoring='r2')
    print(cv_scores)
    print(f'\nr2 mean: {cv_scores.mean()}')
    print(f'std scores: {cv_scores.std()}\n')

    # Calculate and print general metrics
    GeneralMetrics(y_test, y_pred)
    return rf

# Train and evaluate a GradientBoostingRegressor
def GradientBoostingModel(X_train, y_train, X_test, y_test):
    gbr = GradientBoostingRegressor(random_state=1951)

    # Parameter grid for hyperparameter tuning
    param_grid = {
        'n_estimators': [250, 300, 350],
        'max_depth': [3, 5, 7],
        'min_samples_split': [3, 5, 7]
    }
    
    # Perform grid search with cross-validation
    gbr = GridSearchCV(estimator=gbr, param_grid=param_grid, cv=7, n_jobs=-1)

    # Fit the model on training data
    gbr.fit(X_train, y_train)
    print(gbr.best_params_)
    gbr = gbr.best_estimator_

    # Predict on test data
    y_pred = gbr.predict(X_test)

    # Perform cross-validation to evaluate the model
    cv_scores = cross_val_score(gbr, X, y, cv=7, scoring='r2')
    print(cv_scores)
    print(f'\nr2 mean: {cv_scores.mean()}')
    print(f'std scores: {cv_scores.std()}\n')

    # Calculate and print general metrics
    GeneralMetrics(y_test, y_pred)
    return gbr
```

```python
RandomForestModel(X_train, y_train, X_test, y_test)
```

```python
GradientBoostingModel(X_train, y_train, X_test, y_test)
```

### Conclusion
Upon completing the training and checking the metrics in detail, our final model will be GB (Gradient Boosting Model), and we have several reasons.

Primarily, all the errors are lower. The distance between the model's prediction and the real value is smaller, meaning a better approximation between the predicted and actual value.

Additionally, the R value is slightly higher (0.8778 vs. 0.8498). This score ranges from -1 to 1, where a value closer to 1 signifies a better interpretation between the predicted values and the actual values. Essentially, the higher the R value, the better the model interprets when predicting a target (in this case, the price) based on certain features (`gpu`, `ram`, `storage`, etc.).

Finally, we also observe how we conducted cross-validation within the function of each model. This is done to confirm that our model's performance is due to being trained with the necessary characteristics and in the correct manner, not due to a random event. For example, a specific data partition that could lead to overfitting or even a specific partition that causes the model to learn nothing. Cross-validation involves making multiple partitions of our data (7 in this case). In this way, we train with 6 partitions and obtain the metric (R) with the remaining data partition, and so on. By first observing the average, we can note that the cross-validation result of GB is higher. There is also less variation in individual results, with the standard deviation being lower in GB. All this means it is the model with more consistency and robustness if we consider and interpret each metric correctly.
